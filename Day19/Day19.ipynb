```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bestsellers_data = pd.read_csv('datasets/amazon_bestsellers.csv')

bestsellers_data.head()
```

```python
bestsellers_data.describe()
```

```python
bestsellers_data.info()
```

```python
# Visualization to draw the distribution accross a column 
sns.displot(data = bestsellers_data, x = 'Price', aspect = 2)
```

```python
f, ax = plt.subplots(figsize = (12,6))
# This is a figure level plot 
sns.histplot(data = bestsellers_data, x = 'Reviews')
```

```python
bestsellers_data.rename(columns={'User Rating': 'Rating'}, inplace=True)
```

```python
bestsellers_data.head()
```

```python
sns.displot(data = bestsellers_data , x = 'Rating', binwidth = 5, aspect = 2)
```

```python
sns.displot(data = bestsellers_data , x = 'Genre', shrink = 0.5, aspect = 2)
```

```python
sns.displot(data = bestsellers_data, x = 'Price', kind = 'kde', aspect = 2)
```

```python
f, ax = plt.subplots(figsize = (12, 6))
sns.kdeplot(data = bestsellers_data, x = 'Reviews')
```

```python
sns.displot(data = bestsellers_data, x = 'Price', kde= True, aspect= 2)
```

```python
sns.displot(data = bestsellers_data, x = 'Price', kde = True, rug = True, aspect = 2)
```

```python
# obtaining the same result but using a figure graph
f, ax = plt.subplots(figsize = (12, 6))

sns.histplot(data = bestsellers_data, x = 'Price', kde = True)

sns.rugplot(data = bestsellers_data, x = 'Price')
```

```python
# the bandwith smoothing is important for the final representation of the graph
# and its interpretation 
f, ax = plt.subplots(figsize = (12, 6))

sns.kdeplot(data = bestsellers_data, x = 'Price',
           bw_adjust = 0.05, label = 'bw: 0.05'); 

sns.kdeplot(data = bestsellers_data, x = 'Price',
           bw_adjust = 0.2, label = 'bw: 0.2'); 

sns.kdeplot(data = bestsellers_data, x = 'Price',
           bw_adjust = 2, label = 'bw: 2'); 

plt.legend()
```

```python
# the cut parameter removes the extreme values
# from the graph 
f, ax = plt.subplots(figsize = (12, 6))

sns.kdeplot(data = bestsellers_data, x = 'Price',
           bw_adjust = 0.05,cut = 0, label = 'bw: 0.05'); 

sns.kdeplot(data = bestsellers_data, x = 'Price',
           bw_adjust = 0.2,cut = 0, label = 'bw: 0.2'); 

sns.kdeplot(data = bestsellers_data, x = 'Price',
           bw_adjust = 2,cut = 0, label = 'bw: 2'); 

plt.legend()
```
